```python
import os,glob

from Bio import SeqIO
import baltic as bt

base_path='/Users/evogytis/Documents/manuscripts/SARS-CoV-2_kitenis/data/'

ll=bt.loadNexus(os.path.join(base_path,'trees/1ghost.combined.mcc.tre'))
ll.treeStats()

used_seqs=set()
used_seqs=used_seqs.union(set([k.name.split('|')[1] for k in ll.getExternal()]))


ll=bt.loadNexus(os.path.join(base_path,'trees/continent_combined.mcc.tre'))
ll.treeStats()

used_seqs=used_seqs.union(set([k.name.split('|')[1] for k in ll.getExternal()]))


for seq in SeqIO.parse(os.path.join(base_path,'alignments/B.1.620_repr_Cameroon_wRef.fasta'),'fasta'):
    fields=seq.id.split('|')
    if len(fields)>1:
        used_seqs.add(fields[-2])
    else:
        print('not used',seq.id)
        

for seq in SeqIO.parse(os.path.join(base_path,'alignments/B.1.620_wRef_woDRC.fasta'),'fasta'):
    fields=seq.id.split('|')
    if len(fields)>1:
        used_seqs.add(fields[-2])
    else:
        print('not used',seq.id)
```
